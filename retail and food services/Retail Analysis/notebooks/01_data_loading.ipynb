# 📦 Data Loading & Cleaning

```python
import pandas as pd
import os
```

```python
base = r'C:\Users\KRISH\OneDrive\Music\Desktop\retail and food services\retail-analysis'

df = pd.read_csv(base + r'\data\retail.csv')
print("✅ Data Loaded!")
print("Shape:", df.shape)
df.head()
```

```python
print("Columns:", df.columns.tolist())
print("\nData Types:")
print(df.dtypes)
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
df['month'] = pd.to_datetime(df['month'])
df['naics_code'] = pd.to_numeric(df['naics_code'], errors='coerce')
df['value'] = pd.to_numeric(df['value'], errors='coerce')
df.dropna(inplace=True)

print("✅ Data Cleaned!")
print("Shape:", df.shape)
df.head()
```

```python
df.to_csv(base + r'\data\retail_clean.csv', index=False)
print("✅ Saved → retail_clean.csv!")
```
