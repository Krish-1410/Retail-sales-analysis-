# 📊 Exploratory Data Analysis (EDA)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

base = r'C:\Users\KRISH\OneDrive\Music\Desktop\retail and food services\retail-analysis'
df = pd.read_csv(base + r'\data\retail_clean.csv', parse_dates=['month'])
df['year'] = df['month'].dt.year

print("✅ Data Loaded!")
print("Shape:", df.shape)
```

```python
df.describe()
```

```python
top10 = df.groupby('kind_of_business')['value'].sum().nlargest(10)
print(top10)
```

```python
yearly = df.groupby('year')['value'].sum()
print(yearly)
```
